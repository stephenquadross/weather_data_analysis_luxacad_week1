```python
df = pd.read_csv('1. Weather Data.csv')
```

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///my_database.db')
df.to_sql('weather_data', con=engine, if_exists='replace', index=False)
```

```python
%load_ext sql
%sql sqlite:///my_database.db
```

```python
%%sql
SELECT *
FROM weather_data
LIMIT 17;
```

### All records where the weather was exactly clear.

```python
%%sql
SELECT *
FROM weather_data
WHERE Weather = 'Clear';
```

### Number of times the wind speed was exactly 4 km/hr.

```python
%%sql
SELECT *
FROM weather_data
WHERE `Wind Speed_km/h` = 4;
```

### **NULL** values present in the dataset

```python
%%sql

SELECT DISTINCT Weather
FROM weather_data
```

### Renaming the column "Weather" to "Weather_Condition."

```python
%%sql
ALTER TABLE weather_data
RENAME COLUMN Weather TO Weather_Condition;
```

### Mean visibility of the dataset?

```python
%%sql
SELECT AVG(`Visibility_km`) AS Mean_Visibility
FROM weather_data;
```

### Number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km.

```python
%%sql
SELECT COUNT(*) AS Count_Wind24_vis25
FROM Weather_data
WHERE 'Wind Speed_km/h'> 24 AND 'Visibility_km' = 25;
```

### Mean value of each column for each weather condition?

```python
%%sql
SELECT DISTINCT AVG(`Weather_Condition`)
FROM Weather_data;
GROUP BY `Weather_Condition`;
```

Instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.

```python
%%sql
SELECT *
FROM weather_data
WHERE Weather_Condition = 'Clear' AND `Rel Hum_%` > 50 OR `Visibility_km` > 40;
```

### Number of weather conditions that include snow.

```python
%%sql
SELECT COUNT('Weather_Condition') AS Snow_Count
FROM weather_data
WHERE Weather_Condition LIKE '%Snow%';
```
